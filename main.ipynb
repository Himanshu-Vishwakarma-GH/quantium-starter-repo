```python
import pandas as pd
import glob
```

```python
files = glob.glob('data/*.csv')
```

```python
dfs = []
for file in files:
    df = pd.read_csv(file)
    df = df[df['product'] == 'pink morsel'].copy()
    df['sales'] = df['quantity'] * df['price'].str.replace('$', '', regex=False).astype(float)
    df = df[['sales', 'date', 'region']]
    dfs.append(df)
```

```python
df.head()
```

```python
#combine files
final_df = pd.concat(dfs, ignore_index=True)
```

```python
#saving
final_df.to_csv("Pink Morsel Sales.csv", index=False)
```
